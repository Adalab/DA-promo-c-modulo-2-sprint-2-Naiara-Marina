- Extraed los valores únicos su frecuencia. BONUS presenta los resultados en un dataframe. 
    - Donde variable es cada una de las variables categóricas que tenemos en el dataset y n_unicos es una lista con los nombres de los valores únicos de cada variable.

```python
df.columns
```

```python
# para sacar todos y no tener que estar repetiendo el método, hacemos un for loop 
for col in df.columns:
    print("------------------------------------------------------------", "\n")
    print(f"En la columna {col} tenemos los siguientes valores unicos")
    print(pd.DataFrame(df[col].value_counts()))
    print("\n", "------------------------------------------------------------", "\n")
```

```python
# devolver como dataframe del enunciado???
# "Los valores de un diccionario pueden tener cualquier forma: números, strings, listas…"
uniques_freq = {'variable':[], 'n_uniques':[]}
for col in df.columns:
    uniques_freq['variable'].append(col)
    uniques_freq['n_uniques'].append(list(df[col].unique()))

df_unique_freq = pd.DataFrame(uniques_freq)  
df_unique_freq
```
